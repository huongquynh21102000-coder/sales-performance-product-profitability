## 2.1 Các chiều khái niệm và nghiệp vụ (Data Dictionary)
Để đảm bảo các bước tính toán phía sau bám sát thực tế kinh doanh cần phân loại và làm rõ ý nghĩa các cột dữ liệu theo 3 nhóm chính
:** A. Các chiều phân loại dữ liệuiệu.
* `Segment`: Phân khúc khách hàng (Government, Midmarket, Enterprise,...).
* `Country`: Quốc gia/Thị trường diễn ra giao dịch.

* `Discount Band`: Phân loại mức độ chiết khấu (None, Low, Medium, High).

**B. Các chỉ số  với nhau:
* `Units Sold`, `Manufacturing Price`, `Sale Price`: Các chỉ số cơ bản của một đơn vị sản phẩm.

* `Profit`: Lợi nhuận (`Profit` = `Sales` - `COGS`). *Lưu ý: Nếu giá trị bị bọc trong ngoặc đơn `()` thì đó là khoản lỗ.*

**C. Các chiều thời ghân tích xu hướng.
* `Date`, `Month Number`, `Month Name`, `Year`: Thời điểm ghi nhận doanh thu.

* **Lỗi tên cột:** Tên cột có khoảng trắng thừa  

* **Lỗi kiểu dữ liệu:** Các cột tài chính (`Units Sold`, `Gross Sales`, `Discounts`, `Profit`...) lẽ ra phải là dạng tuy nhiên hiện tng là chuỗi văn bản (`object`). Nguyên nhân là do dính các ký hiệu tiền tệ `$`, khoảng trắng và dấu phẩy ngăn cách hàng nghìn.
* **Dữ liệu có định dạng bất thường:** 1. Cột `Discounts` đang dùng ký hiệu `$-` để biểu thị cho số **0**.

> Kế hoạch xử lý
>> 1. **Chuẩn hóa tên cột:** Kiểm tra và xóa bỏ các khoảng trắng bị thừa ở đầu/cuối tên cột (nếu có).
>>2. **Xóa ký tự tiền tệ và dấu trừ:** Loại bỏ dấu đô la (`$`) và dấu gạch ngang (`-`) ra khỏi các cột dữ liệu tài chính.
>>3. **Xóa dấu phẩy phần nghìn:** Bỏ dấu phẩy (`,`) đang nằm bên trong các con số để máy tính không bị nhầm lẫn.

>>5. **Mục tiêu trực quan hóa (Bước tiếp theo):** Sau khi dữ liệu đã sạch và chuẩn, tiến hành vẽ biểu đồ phân tích xu hướng Doanh thu (Sales) và Lợi nhuận (Profit) theo thời gian.ía sau.

```python
# Làm sạch tên cột: Xóa khoảng trắng thừa ở đầu và cuối tên
df.columns = df.columns.str.strip()

print("Tên cột sau khi làm sạch:")
print(df.columns.tolist())
```

### Làm sạch và ép kiểu dữ liệu số

```python
# Tạo danh sách các cột cần làm sạch
cols_to_clean = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales',
                 'Discounts', 'Sales', 'COGS', 'Profit']
```

```python
# duyệt làm sạch qua từng cột
for col in cols_to_clean:
    # 1. Ép tất cả về dạng chuỗi
    df[col] = df[col].astype(str)

    # 2. Xử lý số 0 ($-)
    df[col] = df[col].str.replace('$-', '0', regex=False)

    # 3. Xử lý số âm kế toán (ở cột Profit có dạng $(100) cần chuyển thành -100)
    df[col] = df[col].str.replace('(', '-', regex=False)
    df[col] = df[col].str.replace(')', '', regex=False)

    # 4. Xóa ký hiệu '$', dấu phẩy ',' và khoảng trắng thừa
    df[col] = df[col].str.replace('$', '', regex=False)
    df[col] = df[col].str.replace(',', '', regex=False)
    df[col] = df[col].str.strip()

    # 5. Ép kiểu về số thực (float)
    df[col] = df[col].astype(float)
```

### Doanh thu và tỷ suất lợi nhuận theo thị trường

```python
# 1. Gộp nhóm dữ liệu theo Quốc gia (Thị trường)
#  Tỷ suất lợi nhuận của cả 1 quốc gia phải bằng (Tổng Lợi Nhuận / Tổng Doanh Thu)
market_df = df.groupby('Country').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

# Tạo lại cột Profit Margin (%) chuẩn cho từng quốc gia
market_df['Profit_Margin_%'] = (market_df['Profit'] / market_df['Sales']) * 100

# Tính giá trị trung bình để vẽ đường cắt ngang/dọc chia 4 góc phần tư
mean_sales = market_df['Sales'].mean()
mean_margin = market_df['Profit_Margin_%'].mean()

# 2. Vẽ biểu đồ Phân tán (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=market_df, x='Sales', y='Profit_Margin_%', s=500, color='#FF7F0E', edgecolor='black', alpha=0.8)

# 3. Kẻ 2 đường trung bình chia biểu đồ thành 4 góc phần tư
plt.axvline(x=mean_sales, color='grey', linestyle='--', linewidth=1.5, zorder=0)
plt.axhline(y=mean_margin, color='grey', linestyle='--', linewidth=1.5, zorder=0)

# 4. Gắn tên quốc gia vào từng chấm tròn
for i in range(market_df.shape[0]):
    plt.text(x=market_df['Sales'][i] + 300000, # Cộng thêm 1 chút để chữ không đè lên chấm tròn
             y=market_df['Profit_Margin_%'][i],
             s=market_df['Country'][i],
             fontdict=dict(color='black', size=11, weight='bold'))

# 5. Định dạng biểu đồ
plt.title('Phân tích Tiềm năng Thị trường: Doanh thu vs. Tỷ suất Lợi nhuận', fontsize=15, fontweight='bold')
plt.xlabel('Tổng Doanh thu (Sales - $)', fontsize=12)
plt.ylabel('Tỷ suất lợi nhuận (Profit Margin - %)', fontsize=12)

# Định dạng trục X hiển thị số Triệu ($M) cho dễ nhìn
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${x*1e-6:.1f}M'))

# Đánh dấu các vùng Insight
plt.text(x=market_df['Sales'].min(), y=market_df['Profit_Margin_%'].max(), s='NGÔI SAO MỚI\n(Bán ít - Lãi dày)\n-> Cần Marketing', color='blue', alpha=0.5, weight='bold')
plt.text(x=market_df['Sales'].max() - 4000000, y=market_df['Profit_Margin_%'].min() + 1, s='CẨN TRỌNG\n(Bán nhiều - Lãi mỏng)\n-> Cần tối ưu chi phí', color='red', alpha=0.5, weight='bold')

plt.tight_layout()
plt.grid(True)
plt.show()
```

### Tỷ suất lợi nhuận theo Segment

```python

# 1. Gộp nhóm và tính Tổng Doanh thu, Tổng Lợi nhuận theo Phân khúc
segment_margin_df = df.groupby('Segment').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

# 2. Tính Tỷ suất lợi nhuận (%) thực tế cho từng phân khúc
segment_margin_df['Profit_Margin_%'] = (segment_margin_df['Profit'] / segment_margin_df['Sales']) * 100

# Sắp xếp từ cao xuống thấp để biểu đồ có tính thứ bậc (dễ so sánh)
segment_margin_df = segment_margin_df.sort_values(by='Profit_Margin_%', ascending=False)

# 3. Vẽ biểu đồ Cột (Bar Chart)
plt.figure(figsize=(10, 6))

ax = sns.barplot(data=segment_margin_df, x='Segment', y='Profit_Margin_%', palette='viridis')

# Thêm trực tiếp con số % lên trên đỉnh của từng cột
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f%%', padding=3, fontsize=11, fontweight='bold')

# Định dạng nhãn và tiêu đề
plt.title('Tỷ suất Lợi nhuận (Profit Margin) theo từng Phân khúc', fontsize=15, fontweight='bold')
plt.xlabel('Phân khúc Khách hàng (Segment)', fontsize=12)
plt.ylabel('Tỷ suất Lợi nhuận (%)', fontsize=12)

# Thêm lưới ngang cho dễ dóng thước
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
